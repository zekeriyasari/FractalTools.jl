# Line Attractor 

In this simple notebook, we will consider a simple IFS whose attractor is a simple line, that is, an interval of $[0, 1]$. Consider the following IFS
$$
\begin{aligned} 
    w_1(x) &= \frac{1}{2} x \\
    w_2(x) &= \frac{1}{2} x + \frac{1}{2} \\
\end{aligned}
$$
Let us first construct the IFS

```julia
using FractalTools

ifs = IFS([
Transformation(fill(1 / 2, 1, 1), fill(0, 1)), 
Transformation(fill(1 / 2, 1, 1), fill(1 / 2, 1))
])
```

Note that we constructed two transformations of the form $w(x) = A x + b $ where $A$ is a matrix and $b$ is a vector. Thus, we called `fill(1 / 2, 1, 1)` to construct a $1 \times 1$ matrix. 

Next step is the construction of the attractor. The attractor of this IFS will be computed starting with an initial set of single point, i.e., `[0.25]`

```julia
atr = attractor(ifs, [[0.25]]) 
```

Now let us look deeper into the attractor `atr` 

```julia
@doc Attractor
```

We can now plot the attractor `atr`. Since it is a one dimensional attractor, we plot the attractor set on the `x`-axis

```julia
using Plots
x = getindex.(atr.set, 1)
y = zeros(length(x)) 
scatter(x, y, marker=(:circle, 1))
```
